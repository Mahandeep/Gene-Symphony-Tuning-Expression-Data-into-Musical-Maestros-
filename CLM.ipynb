```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Assuming 'c' is your DataFrame with the specified column names
# gf = c.drop(["var","2","4","8","12","24","48","72"], axis=1)
gf=c.drop(["var","GENE_ID"],axis=1)

# Standardize the data (important for K-means)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(gf)

# Initialize K-means clustering with 100 clusters (adjust the number of clusters as needed)
kmeans = KMeans(n_clusters=100, random_state=100)

# Fit K-means clustering model to the scaled data
kmeans.fit(scaled_data)

# Calculate cluster centroids (means) in original feature space
cluster_centroids = pd.DataFrame(scaler.inverse_transform(kmeans.cluster_centers_), columns=gf.columns)

# Create a new DataFrame with cluster centroids
centroids_df = pd.DataFrame(cluster_centroids.values, columns=gf.columns)

# Add a 'Cluster' column to identify each centroid with its cluster label
centroids_df['Cluster'] = centroids_df.index

# Display the new DataFrame containing cluster centroids
print("Cluster Centroids:")
centroids_df=centroids_df.drop(["Cluster"],axis=1)
centroids_df

pca = PCA(n_components=4)
pca_data = pca.fit_transform(scaled_data)

# Add cluster labels to the PCA-transformed data
pca_df = pd.DataFrame(pca_data, columns=['PC1','PC2','PC3','PC4'])
pca_df['Cluster'] = kmeans.labels_

# Visualize the clusters in 2D PCA space
plt.figure(figsize=(20, 20))
scatter = plt.scatter(pca_df['PC1'],pca_df['PC2'], c=pca_df['Cluster'], cmap='viridis', s=50, alpha=0.8)
plt.colorbar(scatter, label='Cluster')
plt.title('K-means Clustering (2D PCA)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from midiutil import MIDIFile

# Create MIDI file object and add tempo
total_notes = len(centroids_df["0"])  # Total number of notes
notes_per_minute =60  # Number of notes per minute (adjust as needed)
bpm =len(centroids_df["0"])  # Calculate beats per minute based on total notes

# Initialize a MIDIFile object with 7 tracks (0-6)
my_midi_file = MIDIFile(8)

# Add tempo to all tracks (0-6)
for track in range(8):
    my_midi_file.addTempo(track, 0, bpm)  # Use calculated bpm

# Define MIDI program numbers (patch numbers) for each track (channel)
# Define instrument mapping (program numbers) for each track (channel)
instrument_mapping = {
    0: 0,    # Piano
    1: 24,   # Vibraphone
    2: 16,   # Church Organ
    3: 32,   # Acoustic Guitar
    4: 40,   # Bass Guitar
    5: 48,   # Strings
    6: 56,   # Ensemble
    7: 72    # Flute (or another instrument)
}

# Define volume levels for each instrument (adjust as needed)
volume_levels = {
    0: 100,  # Piano
    1: 80,   # Vibraphone
    2: 90,   # Church Organ
    3: 70,   # Acoustic Guitar
    4: 70,   # Bass Guitar
    5: 90,   # Strings
    6: 80,   # Ensemble
    7: 85    # Flute (or another instrument)
}


# Add MIDI notes to each track with different instruments and volumes
for track in range(8):  # Iterate over tracks 0 to 6
    instrument = instrument_mapping[track]  # Get MIDI program number for the current track
    volume = volume_levels[track]  # Get volume level for the current track
    for i in range(total_notes):
        time = (i) * (60 / notes_per_minute)  # Time for this note
        duration = 1  # Duration of the note (in beats, adjust as needed)
        pitch_column = centroids_df.columns[track]
        pitch = centroids_df[pitch_column][i]  # Get pitch value from DataFrame
        if(track==16):
            my_midi_file.addNote(track=track, channel=track, time=time, pitch=pitch, volume=0, duration=duration)
        else:
            my_midi_file.addNote(track=track, channel=track, time=time, pitch=pitch, volume=volume, duration=duration)
            # my_midi_file.addNote(track=track, channel=track, time=time, pitch=pitch, volume=volume//2, duration=duration)
        my_midi_file.addProgramChange(track, channel=track, time=0, program=instrument)  # Set instrument for the track

# Create and save the MIDI file
with open("filename.mid", "wb") as f:
    my_midi_file.writeFile(f)
```
